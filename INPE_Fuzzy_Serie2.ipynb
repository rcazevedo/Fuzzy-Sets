```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function Y = sqrt(1 - (x^2 / a^2))
def fuzzy_function(x, a=2):
    return np.sqrt(1 - (x**2 / a**2))

# Define the fuzzy membership function for set A (example: triangular fuzzy set)
def fuzzy_set_A(x):
    return 1 / (1 + 0.3 * (x - 2)**2)

# Apply extension principle: fuzzify the input
def fuzzify_function(x_values, a=2):
    y_values = fuzzy_function(x_values, a)
    membership_values = fuzzy_set_A(x_values)  # Apply fuzzy membership to x values

    # Apply extension principle: Multiply fuzzy membership values with output values
    fuzzified_y = membership_values * y_values

    return fuzzified_y

# --- Example usage ---
x_values = np.linspace(-2, 2, 400)
fuzzified_output = fuzzify_function(x_values)

# --- Plotting the result for part (a) ---
plt.plot(x_values, fuzzified_output, label="Fuzzified Output (Part a)")
plt.xlabel("x values")
plt.ylabel("Fuzzified Output (Y)")
plt.title("Fuzzification of Function Using Extension Principle (Part a)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the new function Y = sin(x)
def fuzzy_function_sin(x):
    return np.sin(x)

# Apply extension principle with the new function
def fuzzify_function_sin(x_values):
    y_values = fuzzy_function_sin(x_values)
    membership_values = fuzzy_set_A(x_values)  # Use same fuzzy set A

    # Apply extension principle: Multiply fuzzy membership values with output values
    fuzzified_y = membership_values * y_values

    return fuzzified_y

# --- Example usage for (b) ---
fuzzified_output_sin = fuzzify_function_sin(x_values)

# --- Plotting the result for part (b) ---
plt.plot(x_values, fuzzified_output_sin, label="Fuzzified Output (Part b)", color='green')
plt.xlabel("x values")
plt.ylabel("Fuzzified Output (Y)")
plt.title("Fuzzification of Sine Function Using Extension Principle (Part b)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Apply extension principle to both positive and negative x values
def fuzzify_function_both_sides(x_values, a=2):
    y_values = fuzzy_function(x_values, a)
    membership_values = fuzzy_set_A(x_values)  # Apply fuzzy membership to x values

    # Apply extension principle: Multiply fuzzy membership values with output values
    fuzzified_y = membership_values * y_values

    return fuzzified_y

# --- Example usage for (c) ---
x_values_full = np.linspace(-3, 3, 400)  # Include negative values for x
fuzzified_output_full = fuzzify_function_both_sides(x_values_full)

# --- Plotting the result for part (c) ---
plt.plot(x_values_full, fuzzified_output_full, label="Fuzzified Output (Part c)", color='red')
plt.xlabel("x values")
plt.ylabel("Fuzzified Output (Y)")
plt.title("Fuzzification for Both Positive and Negative Inputs Using Extension Principle (Part c)")
plt.legend()
plt.grid(True)
plt.show()
```
