```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function Y = sqrt(1 - (x^2 / a^2))
def fuzzy_function(x, a=2):
    return np.sqrt(1 - (x**2 / a**2))

# Define the fuzzy membership function for set A (example: triangular fuzzy set)
def fuzzy_set_A(x):
    return 1 / (1 + 0.3 * (x - 2)**2)

# Apply extension principle: fuzzify the input
def fuzzify_function(x_values, a=2):
    y_values = fuzzy_function(x_values, a)
    membership_values = fuzzy_set_A(x_values)  # Apply fuzzy membership to x values

    # Apply extension principle: Multiply fuzzy membership values with output values
    fuzzified_y = membership_values * y_values

    return fuzzified_y

# --- Example usage ---
x_values = np.linspace(-2, 2, 400)
fuzzified_output = fuzzify_function(x_values)

# --- Plotting the result for part (a) ---
plt.plot(x_values, fuzzified_output, label="Fuzzified Output (Part a)")
plt.xlabel("x values")
plt.ylabel("Fuzzified Output (Y)")
plt.title("Fuzzification of Function Using Extension Principle (Part a)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funo quadrtica para calcular Z = x^2 + y^2
def fuzzy_function(x, y, a=2):
    return np.sqrt(1 - (x**2 / a**2) - (y**2 / a**2))

# Funo de pertinncia fuzzy para o conjunto A
def fuzzy_set_A(x, y):
    return 1 / (1 + 0.3 * ((x - 2)**2 + (y - 2)**2))

# Gerar a superfcie fuzzy
def fuzzify_surface(x_vals, y_vals, a=2):
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = fuzzy_function(X, Y, a)  # Calcular a superfcie
    membership = fuzzy_set_A(X, Y)  # Aplicar a funo de pertinncia fuzzy

    # Aplicando o princpio da extenso: multiplicar a funo de pertinncia pela superfcie
    fuzzified_Z = membership * Z
    return X, Y, fuzzified_Z

# Definir intervalos para X e Y
x_vals = np.linspace(-2, 2, 400)
y_vals = np.linspace(-2, 2, 400)

# Calcular os valores da superfcie fuzzificada
X, Y, fuzzified_Z = fuzzify_surface(x_vals, y_vals)

# Plotar a superfcie quadrtica
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, fuzzified_Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Fuzzified Output')
ax.set_title('Fuzzification of Quadratic Surface Using Extension Principle (Part a)')

# Exibir o grfico
plt.show()

# Parte (a) - Clculo do grfico resultante
def part_a_solution():
    print("Exerccio 11 - Parte (a): Fuzzificao com Funo Quadrtica:")
    print(f"Resultado da superfcie fuzzificada gerada para a funo quadrtica.")

part_a_solution()
```
