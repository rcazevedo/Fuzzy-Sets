```python
import matplotlib.pyplot as plt
import numpy as np

# Definicao dos intervalos (Cortes-Alfa)
alphas = [0.33, 0.67, 1.0]
A_intervals = [(6, 10), (7, 9), (8, 8)]
B_intervals = [(1, 5), (2, 4), (3, 3)]

# Calculo de C = A - B -> [a_min - b_max, a_max - b_min]
C_intervals = []
for a, b in zip(A_intervals, B_intervals):
    c_min = a[0] - b[1]
    c_max = a[1] - b[0]
    C_intervals.append((c_min, c_max))

print("--- Resultado Problema 1 (C = A - B) ---")
for alpha, interval in zip(alphas, C_intervals):
    print(f"Alpha {alpha:.2f}: [{interval[0]:.1f}, {interval[1]:.1f}]")

# Plotagem do Esboco
x_vals = [1, 3, 5, 7, 9]
y_vals = [0.33, 0.67, 1.0, 0.67, 0.33]

plt.figure(figsize=(8, 4))
plt.plot(x_vals, y_vals, 'bo-', label='C = A - B')
plt.title('Problema 1: Subtracao de Numeros Fuzzy')
plt.xlabel('Universo de Discurso')
plt.ylabel('Pertinencia')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# (Reutilizando os intervalos definidos no Problema 1)

# Calculo de C = A * B -> [a_min * b_min, a_max * b_max] (Positivos)
C_mult = []
for a, b in zip(A_intervals, B_intervals):
    c_min = a[0] * b[0]
    c_max = a[1] * b[1]
    C_mult.append((c_min, c_max))

print("\n--- Resultado Problema 2 (C = A * B) ---")
for alpha, interval in zip(alphas, C_mult):
    print(f"Alpha {alpha:.2f}: [{interval[0]:.1f}, {interval[1]:.1f}]")

# Dados para plotagem
x_mult = [6, 14, 24, 36, 50]
y_mult = [0.33, 0.67, 1.0, 0.67, 0.33]

plt.figure(figsize=(8, 4))
plt.plot(x_mult, y_mult, 'ro-', label='C = A * B')
plt.title('Problema 2: Multiplicacao de Numeros Fuzzy')
plt.xlabel('Universo de Discurso')
plt.ylabel('Pertinencia')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculo de C = A / B -> [a_min / b_max, a_max / b_min]
C_div = []
for a, b in zip(A_intervals, B_intervals):
    c_min = a[0] / b[1]
    c_max = a[1] / b[0]
    C_div.append((c_min, c_max))

print("\n--- Resultado Problema 3 (C = A / B) ---")
for alpha, interval in zip(alphas, C_div):
    print(f"Alpha {alpha:.2f}: [{interval[0]:.2f}, {interval[1]:.2f}]")

# Dados para plotagem (aproximados para visualizacao)
x_div = [1.2, 1.75, 2.67, 4.5, 10]
y_div = [0.33, 0.67, 1.0, 0.67, 0.33]

plt.figure(figsize=(8, 4))
plt.plot(x_div, y_div, 'go-', label='C = A / B')
plt.title('Problema 3: Divisao de Numeros Fuzzy')
plt.xlabel('Universo de Discurso')
plt.ylabel('Pertinencia')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Definicao dos conjuntos discretos A e B (Simulando Exemplo 4.2)
A = {1:0.3, 2:0.7, 3:1.0, 4:0.7, 5:0.3}
B = {5:0.2, 6:0.6, 7:1.0, 8:0.6, 9:0.2}

Z = {} # Resultado da subtracao

# Aplicacao do Principio da Extensao
for x, mu_x in A.items():
    for y, mu_y in B.items():
        z = x - y
        term_mu = min(mu_x, mu_y) # Intersecao (Min)

        # Uniao (Max)
        if z in Z:
            Z[z] = max(Z[z], term_mu)
        else:
            Z[z] = term_mu

print("\n--- Resultado Problema 4 (Extensao: A - B) ---")
# Ordenar chaves para impressao
for z_val in sorted(Z.keys()):
    print(f"z = {z_val}: mu = {Z[z_val]:.2f}")

# Plotagem
lists = sorted(Z.items())
x_ext, y_ext = zip(*lists)

plt.figure(figsize=(8, 4))
plt.bar(x_ext, y_ext, width=0.4, label='Z = A - B', color='purple')
plt.title('Problema 4: Subtracao via Principio da Extensao')
plt.xlabel('z')
plt.ylabel('Pertinencia')
plt.grid(True, axis='y')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Funcoes lambda para os intervalos
# A: Triangular (8, 18, 32)
func_A = lambda a: (10*a + 8, 32 - 14*a)
# B: Triangular (-3, 6, 24)
func_B = lambda a: (9*a - 3, 24 - 18*a)

def multiply_robust(I1, I2):
    p = [I1[0]*I2[0], I1[0]*I2[1], I1[1]*I2[0], I1[1]*I2[1]]
    return min(p), max(p)

alphas = np.linspace(0, 1, 11)
rows = []

for a in alphas:
    A = func_A(a)
    B = func_B(a)

    # Soma
    C = (A[0] + B[0], A[1] + B[1])

    # Subtracao
    D = (A[0] - B[1], A[1] - B[0])

    # Multiplicacao
    E = multiply_robust(A, B)

    rows.append([a,
                 f"[{C[0]:.1f}, {C[1]:.1f}]",
                 f"[{D[0]:.1f}, {D[1]:.1f}]",
                 f"[{E[0]:.1f}, {E[1]:.1f}]"])

print("\n--- Resultado Problema 5 (Grid de Intervalos) ---")
df = pd.DataFrame(rows, columns=["Alpha", "Soma", "Subtracao", "Mult"])
print(df.to_string(index=False))

# --- Plotagem da Soma (C) ---
C_L = [func_A(a)[0] + func_B(a)[0] for a in alphas]
C_R = [func_A(a)[1] + func_B(a)[1] for a in alphas]

plt.figure(figsize=(8, 5))
plt.plot(C_L, alphas, 'b-', label='Lado Esquerdo')
plt.plot(C_R, alphas, 'b-', label='Lado Direito')
plt.hlines(0, min(C_L), max(C_R), colors='k', linestyles='dotted')
plt.title('Problema 5a: Adicao A + B')
plt.xlabel('x')
plt.ylabel('Alpha')
plt.grid(True)
plt.legend()
plt.show()

# --- Plotagem da Subtracao (D) ---
D_L = [func_A(a)[0] - func_B(a)[1] for a in alphas]
D_R = [func_A(a)[1] - func_B(a)[0] for a in alphas]

plt.figure(figsize=(8, 5))
plt.plot(D_L, alphas, 'r-', label='Lado Esquerdo')
plt.plot(D_R, alphas, 'r-', label='Lado Direito')
plt.hlines(0, min(D_L), max(D_R), colors='k', linestyles='dotted')
plt.title('Problema 5b: Subtracao A - B')
plt.xlabel('x')
plt.ylabel('Alpha')
plt.grid(True)
plt.legend()
plt.show()

# --- Plotagem da Multiplicacao (E) ---
E_min = [multiply_robust(func_A(a), func_B(a))[0] for a in alphas]
E_max = [multiply_robust(func_A(a), func_B(a))[1] for a in alphas]

plt.figure(figsize=(8, 5))
plt.plot(E_min, alphas, 'g-', label='Lado Esquerdo')
plt.plot(E_max, alphas, 'g-', label='Lado Direito')
plt.hlines(0, min(E_min), max(E_max), colors='k', linestyles='dotted')
plt.title('Problema 5c: Multiplicacao A * B (Forma Nao-Linear)')
plt.xlabel('x')
plt.ylabel('Alpha')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Aproximacao dos intervalos para alpha=0.6
# A: [1, 4]
# B: Suporte ~[0, 4], interpolando para 0.6 -> aprox [0.5, 3.5] (Exemplo)
# C: Suporte ~[1, 4], interpolando para 0.6 -> aprox [1, 3.5] (Exemplo)

# Vamos usar intervalos ilustrativos baseados nos suportes dados
# A_0.6
a_L, a_R = 1, 4
# B_0.6 (Interpolacao visual)
b_L, b_R = 0.5, 3.5
# C_0.6 (Interpolacao visual)
c_L, c_R = 1, 3.5

# LHS: A * (B + C)
sum_bc_L = b_L + c_L
sum_bc_R = b_R + c_R
lhs_L = a_L * sum_bc_L
lhs_R = a_R * sum_bc_R

# RHS: (A * B) + (A * C)
ab_L = a_L * b_L
ab_R = a_R * b_R
ac_L = a_L * c_L
ac_R = a_R * c_R
rhs_L = ab_L + ac_L
rhs_R = ab_R + ac_R

print("\n--- Resultado Problema 6 (Verificacao em Alpha ~ 0.6) ---")
print(f"LHS [A(B+C)]: [{lhs_L:.2f}, {lhs_R:.2f}]")
print(f"RHS [AB+AC]:  [{rhs_L:.2f}, {rhs_R:.2f}]")

if (rhs_L <= lhs_L) and (rhs_R >= lhs_R):
    print("Resultado: A sub-distributividade (LHS contido em RHS) foi observada.")
    print("O RHS e mais largo devido a dependencia da variavel A.")
else:
    print("Resultado inconclusivo para os valores testados.")
```
